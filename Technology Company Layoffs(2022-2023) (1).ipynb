# 1.Proportion of Each Year

**Shows each year’s contribution to total layoffs.**

* In 2022, layoffs started increasing as tech companies realized they had over-hired during the pandemic.

* In 2023, the layoffs peaked — companies like Google, Amazon, Meta, and Microsoft cut tens of thousands of jobs.

* Reasons included cost cutting, economic slowdown, and AI automation replacing some roles.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert and extract year
data["reported_date"] = pd.to_datetime(data["reported_date"], errors="coerce")
data["year"] = data["reported_date"].dt.year

# Keep only 2022 and 2023 data
data_filtered = data[data["year"].isin([2022, 2023])]

# Ensure numeric layoffs
data_filtered["total_layoffs"] = pd.to_numeric(data_filtered["total_layoffs"], errors="coerce")

# Group total layoffs by year
layoffs_by_year = data_filtered.groupby("year")["total_layoffs"].sum()

# Plot line chart
plt.figure(figsize=(8,5))
sns.lineplot(
    x=layoffs_by_year.index,
    y=layoffs_by_year.values,
    marker="o",
    linewidth=3,
    color="dodgerblue"
)

plt.title("Technology Company Layoffs Trend (2022–2023)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Total Layoffs")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## Insights

Layoffs rose sharply in 2022 as tech companies recognized pandemic over-hiring and began correcting workforce sizes, but they peaked in 2023, driven by massive cuts from Google, Amazon, Meta, Microsoft, and others. This surge was fueled by cost-cutting measures, global economic slowdown, and the rapid adoption of AI and automation, leading companies to restructure and reduce roles that were no longer essentia

# 2.Industry-wise layoffs

**Causes:**

* Too much hiring during COVID time.

* Less investor money and high interest rates.

* Government rules became strict.

**Improvements:**

* Spend money carefully.

* Follow rules properly.

* Create new useful financial products.

```python
industry_layoffs = (
    data.groupby("industry")["total_layoffs"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
plt.figure(figsize=(10,5))
plt.barh(industry_layoffs.index, industry_layoffs.values, color="steelblue")
plt.title("Top 10 Industries by Total Layoffs (2022–2023)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## Insights

Industries that hired heavily during the COVID boom faced the highest layoffs as investor funding decreased, interest rates rose, and regulatory pressures increased. These layoffs show that industries must manage hiring more responsibly, follow compliance requirements, and innovate with better financial or digital products to maintain long-term stability.

# 4.Layoffs by Company Type (Public vs Private) — 2022–2023)

* Compares total layoffs between public and private tech companies.

* Typically, private companies have more layoffs because of limited funds,

* while public companies cut jobs mainly for cost control.

```python
# Create a simple public/private classification
public_companies = ["Google", "Meta", "Amazon", "Microsoft", "Apple", "Salesforce", "IBM", "Intel", "Netflix"]
data["company_type"] = data["company"].apply(
    lambda x: "Public" if any(pub.lower() in str(x).lower() for pub in public_companies) else "Private"
)

# Group by company type and calculate total layoffs
layoffs_by_type = data.groupby("company_type")["total_layoffs"].sum().sort_values(ascending=False)

# Plot bar graph
plt.figure(figsize=(7,5))
bars = plt.bar(layoffs_by_type.index, layoffs_by_type.values, color=["orange", "teal"])
plt.title("Layoffs by Company Type (Public vs Private) — 2022–2023")
plt.xlabel("Company Type")
plt.ylabel("Total Layoffs")

# Add value labels
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f"{int(bar.get_height())}", ha="center", va="bottom", fontsize=10)

plt.tight_layout()
plt.show()
```

## Insights

Public companies reported larger layoffs in 2022–2023 due to aggressive cost-cutting, over-hiring during the pandemic, and pressure from shareholders to maintain profitability, while private companies showed fewer but more survival-driven layoffs caused mainly by funding shortages and declining investor support.
